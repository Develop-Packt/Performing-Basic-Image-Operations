```python
# Import modules
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img1 = cv2.imread("../data/board.png")
img2 = cv2.imread("../data/board2.png")
```

```python
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(img1,cmap="gray")
plt.show()
```

```python
plt.imshow(img2,cmap="gray")
plt.show()
```

```python
# Set threshold and maximum value
thresh = 150
maxValue = 255
```

```python
# Binary threshold
th, dst1 = cv2.threshold(img1, thresh, maxValue, cv2.THRESH_BINARY)
```

```python
# Binary threshold
th, dst2 = cv2.threshold(img2, thresh, maxValue, cv2.THRESH_BINARY)
```

```python
plt.imshow(dst1, cmap='gray')
plt.show()
```

```python
plt.imshow(dst2, cmap='gray')
plt.show()
```

```python
dst = cv2.bitwise_xor(dst1,dst2)
```

```python
plt.imshow(dst, cmap='gray')
plt.show()
```
