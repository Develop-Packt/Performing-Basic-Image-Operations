```python
# Import libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Read image
img = cv2.imread("../data/puppy.jpg")
```

```python
# Display image
plt.imshow(img[:,:,::-1])
plt.show()
```

```python
cvImg = cv2.multiply(img,2)
```

```python
plt.imshow(cvImg[:,:,::-1])
plt.show()
```

```python
npImg = img*2
```

```python
plt.imshow(npImg[:,:,::-1])
plt.show()
```
