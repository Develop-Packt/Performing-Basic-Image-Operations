```python
# Import modules
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread("../data/zebra.jpg")
```

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(img, cmap='gray')
plt.show()
```

```python
# Set threshold and maximum value
thresh = 150
maxValue = 255
```

```python
# Binary threshold
th, dst = cv2.threshold(img, thresh, maxValue, cv2.THRESH_BINARY)
```

```python
plt.imshow(dst, cmap='gray')
plt.show()
```
