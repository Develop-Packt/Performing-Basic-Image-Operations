```python
# Import modules
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread("../data/book.jpg")
```

```python
plt.imshow(img[:,:,::-1])
plt.show()
```

```python
inputPts = np.float32([[4,381],
                      [266,429],
                      [329,68],
                      [68,20]])
outputPts = np.float32([[0,300],
                       [300,300],
                       [300,0],
                       [0,0]])
```

```python
M = cv2.getPerspectiveTransform(inputPts,outputPts)
```

```python
dst = cv2.warpPerspective(img,M,(300,300))
```

```python
plt.imshow(dst[:,:,::-1])
plt.show()
```
